Load Model

```python
from datasets import Dataset
mnist_model = Dataset()
# mnist.data_frame
```

Visualize examples

```python
mnist_model.render_visualization_data()
```

Division of the data set

```python
from sklearn.model_selection import train_test_split
# divide the model for test and train
INPUT_train, INPUT_test, OUTPUT_train, OUTPUT_test = train_test_split(mnist_model.mnist.data, mnist_model.mnist.target, test_size=0.3)
print(len(OUTPUT_train))
print(len(OUTPUT_test))
```

Training Algorithm

```python
from sklearn.neural_network import MLPClassifier

"""
hidden layer: number of layers beetween OUTPUT AND INPUT layers
activation function: the sigmoid function
solver: the optimization weight, gradient descent.
"""
clf = MLPClassifier(hidden_layer_sizes=(50, ), activation="logistic", solver="sgd")
clf.fit(INPUT_train, OUTPUT_train)
```

```python
clf.n_layers_
```

Test Set for prediction

```python
output_predict = clf.predict(INPUT_test)
```

```python
from sklearn.metrics import f1_score
f1_score(OUTPUT_test, output_predict, average="weighted")
```
